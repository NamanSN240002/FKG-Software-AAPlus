#Collabarative filtering based Recommender System


```python
x = ratings_with_name.groupby('user_ssn').count()['user_rating'] > 0
padhe_likhe_users = x[x].index
```

```python
final_ratings = filtered_rating[filtered_rating['product_name'].isin(popular_products)]
```

```python
pt = final_ratings.pivot_table(index='product_name',columns='user_ssn',values='user_rating')
```

#Recommending based on product similarity

```python
def recommend(product_name):
    # index fetch
    index = np.where(pt.index==product_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]

    data = []
    for i in similar_items:
        item = []
        temp_df = products[products['product_name'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('product_name')['product_name'].values))
        item.extend(list(temp_df.drop_duplicates('product_name')['brand'].values))
        item.extend(list(temp_df.drop_duplicates('product_name')['image'].values))


        data.append(item)

    return data
```

#Collabarative filtering on User similarity

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from scipy import sparse
```

```python

# Load the purchase history data into a DataFrame
data = ratings
purchase_history = pd.DataFrame(data)

# Count the number of purchases for each user and product combination
purchase_counts = purchase_history.groupby(['user_ssn', 'product_id']).size().unstack(fill_value=0)

# Convert the purchase counts to a sparse matrix
sparse_purchase_counts = sparse.csr_matrix(purchase_counts)

# Compute the cosine similarity matrix between the products
cosine_similarities = cosine_similarity(sparse_purchase_counts.T)
```

```python
purchase_counts.head()
```

#Recommending products based on user purchase patterns

```python
# Define a function to recommend items for a user based on their purchase history
def recommend_items(user_ssn, n=10):
    user = -1
    for index, row in users.iterrows():
      if row['ssn']==user_ssn:
        user = index
        break

    # Get the user's purchase history
    user_history = sparse_purchase_counts[user].toarray().flatten()

    # Compute the average cosine similarity between the user's purchased items and all other items
    similarities = cosine_similarities.dot(user_history)

    # Get the indices of the user's purchased items
    purchased_indices = np.where(user_history > 0)[0]

    # Set the similarity scores for purchased items to 0
    similarities[purchased_indices] = 0

    # Sort the items by similarity score and return the top n items
    recommended_indices = np.argsort(similarities)[::-1][:n]
    recommended_items = list(purchase_counts.columns[recommended_indices])

    # Remove the items that the user has already purchased
    purchased_items = list(purchase_counts.columns[purchase_counts.loc[user_ssn] > 0])
    recommended_items = [item for item in recommended_items if item not in purchased_items]

    return recommended_items

```

```python
from flask import Flask
from flask_restful import abort,Api, Resource, reqparse

app = Flask(__name__)
api = Api(app)
```

```python
register_put_args = reqparse.RequestParser()
register_put_args.add_argument("name",type=str,help="Name is required",required=True)
register_put_args.add_argument("ssn",type=str,help="Ssn is required",required=True)
register_put_args.add_argument("password1",type=str,help="Password is empty",required=True)
register_put_args.add_argument("password2",type=str,help="Confirm Password is empty",required=True)
register_put_args.add_argument("sex",type=str,help="Sex is required",required=True)
register_put_args.add_argument("job",type=str,help="Job is required",required=True)
register_put_args.add_argument("company",type=str,help="Company is required",required=True)
register_put_args.add_argument("dob",type=str,help="Date of Birth is required",required=True)

login_get_args = reqparse.RequestParser()
login_get_args.add_argument("ssn",type=str,help="Ssn is required",required=True)
login_get_args.add_argument("password",type=str,help="Password is required",required=True)

top_recommendation_post_args = reqparse.RequestParser()
top_recommendation_post_args.add_argument("user_ssn",type=str,help="Enter ssn!",required=True)

product_recommendation_post_args = reqparse.RequestParser()
product_recommendation_post_args.add_argument("products",type=str,help="Enter products!",required=True)
```

```python
class Register(Resource):
    def put(self):
        d = {}
        args = register_put_args.parse_args()
        ssn = args['ssn']
        name = args['name']
        sex = args['sex']
        job = args['job']
        company = args['company']
        dob = args['dob']
        password1 = args['password1']
        password2 = args['password2']
        if password1!=password2 :
            abort(400,custom="Passowrds don't match!")
        if (users['ssn']==ssn).any():
            return {"status":"Error","msg":"User already exists!"}
        to_add = {
            "ssn":ssn,
            "name":name,
            "sex":sex,
            "company":company,
            "birthdate":dob,
            "job":job,
            "password":password1
        }
        users.append(to_add,ignore_index=True)
        users.to_csv('user.csv')
        d['output'] = to_add
        d['status'] = "OK"
        return d
```

```python
class Login(Resource):
    def post(self):
        args = login_get_args.parse_args()
        ssn = args['ssn']
        password = args['password']
        if not (users['ssn']==ssn).any():
            return {"Status":"Error","msg":"User Not Found!"}
        user = users.groupby('ssn').get_group(ssn)
        if (user['password']==password).any():
            return {
            "Status":"OK",
            "msg":"Login Succesful!",
            "user":{
                    "ssn":str(user['ssn']).split()[1],
                    "name":str(user['name']).split()[1],
                    "sex":str(user['sex']).split()[1],
                    "job":str(user['job']).split()[1],
                    "company":str(user['company']).split()[1],
                    "Date of Birth":str(user['birthdate']).split()[1]
                }
            }
        return {"Status":"Error","msg":"Wrong Password!"}
```

```python
class TopRecommendation(Resource):
    def post(self):
        args = top_recommendation_post_args.parse_args()
        user_ssn = args['user_ssn']
        prod_id_list = recommend_items(user_ssn)
        prod_id_dict = []
        for index,row in products.iterrows():
            for j in prod_id_list:
                if row['product_id']==j:
                    prod_id_dict.append({
                        "product_name":row['product_name'],
                        "image":row['image'],
                        "discounted_price":row['discounted_price'],
                        "description":row['description'],
                        "url":row['product_url']
                    })

        return {"data":prod_id_dict}

class ProductRecommendation(Resource):
    def post(self):
        args = product_recommendation_post_args.parse_args()
        products = args['products']
        prod_id_list = recommend(products)
        return {"data":prod_id_list}
```

```python
api.add_resource(TopRecommendation,'/top_recommendation')
api.add_resource(ProductRecommendation,'/product_recommendation')
api.add_resource(Register,'/register')
api.add_resource(Login,'/login')
```

```python
import regex as re
```

```python
def get_nth_category(level=1):
    """extract the level-n product category from the product category tree"""

    if level == 1:
        category = products.product_category_tree.apply(lambda x: re.split(" >> ", x)[0]).str[2:]
    else:
        category = products.product_category_tree.apply(lambda x: re.split(" >> ", x)[level:(level+1)])

    category = category.replace("[]", "[EMPTY_LEVEL]")

    return category
```
